# SIB Keyboard - example data
## Load data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'example_log.csv'

x_names = ['L%d%c' % (i, c) for i in range(21) for c in ['x', 'y', 'z']] 
y_names = ['a', 's', 'd', 'f']

col_names = x_names + y_names

df = pd.read_csv(file_path, names=col_names)
```

```python
df[x_names]
```

## xy-value plot

```python
df = df.iloc[df.index[100:-100]]
```

```python
x_df = df[x_names]

x_means = x_df.mean()
```

```python
xs = x_means[['L%dx' % i for i in range(21)]]
ys = x_means[['L%dy' % i for i in range(21)]]
zs = x_means[['L%dz' % i for i in range(21)]]
```

```python
plt.figure(figsize=(10, 10))

for i in range(21):
    plt.plot(ys[i], xs[i], 'r.')
    plt.text(ys[i], xs[i], '%d' % i)

plt.show()
```

## Z-value plot

```python
x_df = df[x_names].iloc[:200]

x_means = x_df.mean()
x_stds  = x_df.std()

df[x_names] = (df[x_names] - x_means) / x_stds
```

```python
df = df.iloc[df.index[:340]]
# df = df.iloc[df.index[500:700]]
```
