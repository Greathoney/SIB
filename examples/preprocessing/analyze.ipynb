```python
MIN_BOUND = -6.420232e+00
MAX_BOUND = 1.670866e+00

def normalize(h):
    h = (h - MIN_BOUND) / (MAX_BOUND - MIN_BOUND)
    h[h>1] = 1.
    h[h<0] = 0.
    return h

i = normalize(h)
```
