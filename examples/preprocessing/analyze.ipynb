```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
file_path = "/home/hwang/SIB/examples/preprocessing/log.csv"


x_names = ['L%d%c' % (i, c) for i in range(21) for c in ['x', 'y', 'z']] 
y_names = ['R%d%c' % (i, c) for i in range(21) for c in ['x', 'y', 'z']] 
z_names = ['a', 's', 'd', 'f']

col_names = x_names + y_names + z_names

df = pd.read_csv(file_path,names=col_names)
```

```python
a = df['R0x']
b = pd.DataFrame(a)
```

```python
plt.figure(figsize=(5, 5))

plt.plot(b)
plt.show()
```

```python
scaler = StandardScaler()
c = scaler.fit_transform(b)
```

```python
scaler = RobustScaler()
d = scaler.fit_transform(b)
```

```python
plt.figure(figsize=(10,5))
plt.plot(c)
plt.plot(d)
plt.show()
```
