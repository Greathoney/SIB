```python
import numpy as np
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = '/home/kju/SIB/dataset/example_log.csv'

x_names = ['L%d%c' % (i, c) for i in range(21) for c in ['x', 'y', 'z']] 
y_names = ['a', 's', 'd', 'f']

col_names = x_names + y_names

df = pd.read_csv(file_path, names=col_names)
```

```python
df = df.iloc[df.index[:340]]
df[x_names]
```

```python
# plt.plot(df['L0x'])
plt.hist(df['L0x'])
plt.show()
```

```python

# print(df[x_names].min())

scaler = RobustScaler()
dff = scaler.fit_transform(df[x_names])

plt.hist(dff[0])
plt.show()

# print(df[x_names].max())
```

```python
# print(scaler.center_) # median of df
# print(scaler.scale_)

print('df[x_names] : ',df.shape)
print('dff : ',dff.shape)
print("스케일 조정 전 특성별 최소:\n", df[x_names].min(axis=0))
print("스케일 조정 후 특성별 최소:\n", dff.min(axis=0))
print("스케일 조정 전 특성별 최대:\n", df[x_names].max(axis=0))
print("스케일 조정 후 특성별 최대:\n", dff.max(axis=0))

# print(np.median(dff[0])) #+df['L0x'].median()
# print(np.median(df['L0x']) )
```
